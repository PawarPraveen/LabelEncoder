```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("house_price_dataset.csv")
df
```

```python
df.isnull().sum()
```

```python
df["Bedrooms"] = df["Bedrooms"].fillna(df["Bedrooms"].median())
```

```python
df.Bathrooms = df.Bathrooms.fillna(df["Bathrooms"].median())
```

```python
data_impute = ['Bedrooms' , 'Bathrooms', 'Garage']

for data in data_impute:
    median_value = df[data].median()
    print(f"filling missing value in'{data}' with median {median_value}")
    df[data] = df[data].fillna(median_value)
```

```python
df
```

```python
from sklearn.preprocessing import LabelEncoder
ft = LabelEncoder()
df["City_Location"] = ft.fit_transform(df["Location"])
```

```python
from sklearn import linear_model
model = linear_model.LinearRegression()
```

```python
model.fit(df[['Area','Bedrooms','Bathrooms', 'Floors','YearBuilt', 'Garage', 'City_Location']], df.Price)
```

```python
model.predict([[2200,4,3,2,2010,2,2]])
```
